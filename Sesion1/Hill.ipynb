```python
from glob import glob
rutaLimpio = './Textos/*.txt'
listaLibros = glob(rutaLimpio)
listaLibros
```

```python
textos = []
for libro in listaLibros:
    texto = ""
    for char in open(libro).read():
        if char.isalpha():
            texto += char

    texto = texto.upper()
    textos.append(texto)
```

```python
from copy import deepcopy
freqs = []
for texto in textos:
    freq = {}
    
    for i in range (0, len(texto) + 3, 3):
        if texto[max(0, i-3):i] in freq:
            freq[texto[max(0, i-3):i]] += 1
        else:
            freq[texto[max(0, i-3):i]] = 1
            
    sortedD = {}
    sortedK = sorted(freq, key=freq.get, reverse=True)  # [1, 3, 2]

    for w in sortedK:
        sortedD[w] = freq[w]

    freqs.append(deepcopy(sortedD)) # {1: 1, 3: 4, 2: 9}
```

```python
letra = list(sorted_dict.keys())[0]
repes = list(sorted_dict.values())[0]
print(letra + " " + str(repes))
```

```python
for index, f in enumerate(freqs):
    letra = list(f.keys())[0]
    repes = list(f.values())[0]
    print("Indice " + str(index) + " letra " + letra + " " + str(repes))
```
