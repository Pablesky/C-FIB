```python
import os
actual = os.getcwd()
actual
```

```python
RTEXTO = actual + '/descarga'
RTEXTOS = actual + '/Textos/*.txt'
RPADDLE = actual + '/Paddle/'
```

```python
from glob import glob
LTEXTOS = glob(RTEXTOS)
LTEXTOS;
```

```python
FTEXTO = open(RTEXTO)
TEXTO = FTEXTO.read()
TEXTO
```

```python
textoNum = []
for char in TEXTO:
    if not char in ['\n', '\ufeff']:
        print(char, end='')
        print(' ' + str(ord(char)))
        textoNum.append(ord(char))
```

```python
listaAjustada = []
for elem in textoNum:
    if elem > 9000:
        listaAjustada.append(elem)

for elem in listaAjustada:
    print(chr(elem), end=' ')
    print(elem)
```

```python
for paddle in range(0, 26):
    file = open(RPADDLE + 'Paddle' + str(paddle) + '.txt', 'a')
    file = open(RPADDLE + 'Paddle' + str(paddle) + '.txt', 'w')
    for elem in listaAjustada:
        file.write(str(elem - paddle) + '\n')
```
