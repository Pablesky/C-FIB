# Escitala
Cambiar ruta original *rutaOriginal* por el texto deseado.

Cambiar *longitudLinia* por la largada que quieras imprimir de cada iteración.

```python
rutaOriginal = 'EscitalaExamen'
fichero = open(rutaOriginal)
texto = fichero.read()
fichero.close
print(texto[0:20])
```

```python
longitudLinia = 50
for filas in range(1, 200):
    print("Vuelta en " + str(filas) + '\n')
    tabla = []
    for i in range(0, filas):
        tabla.append([])

    indice = 0
    for char in texto:
        tabla[indice].append(char)
        indice += 1
        indice = indice % filas
    
    linia = ""
    for char in tabla[0]:
        linia += char
    print(linia[0:longitudLinia])
```
