*freqTri* los trigramas en ingles m√°s usados.

*rutaOriginal* es el nombre del fichero a leer.

```python
import sympy as sp
from sympy import *
from itertools import permutations
```

```python
modulo = 26
```

```python
#freqTri = ['the','and','ing','her','hat','his','tha','ere','for','ent','ion','ter','was','you','ith','ver','all','wit','thi','tio']
freqTri = ['THE','AND','ING','HER','HAT','HIS','THA','ERE','FOR','ENT','ION','TER','WAS','YOU','ITH','VER','ALL','WIT','THI','TIO']
```

```python
rutaOriginal = 'Hill'
fichero = open(rutaOriginal)
texto = fichero.read()
fichero.close
print(texto[0:20])
```

```python
frecuencias = {}
for i in range (0, len(texto) + 3, 3):
    if texto[max(0, i-3):i] in frecuencias:
        frecuencias[texto[max(0, i-3):i]] += 1
    else:
        frecuencias[texto[max(0, i-3):i]] = 1
```

```python
sortedD = {}
sortedK = sorted(frecuencias, key=frecuencias.get, reverse=True)  # [1, 3, 2]

for w in sortedK:
    sortedD[w] = frecuencias[w]

frecuencias = sortedD
print(frecuencias)
```

```python
lista3primeras = list(frecuencias.keys())[:3]
[primera, segunda, tercera] = lista3primeras
print(primera)
print(segunda)
print(tercera)
```

```python
encriptado = Matrix([[ord(primera[0]), ord(segunda[0]), ord(tercera[0])],
                     [ord(primera[1]), ord(segunda[1]), ord(tercera[1])],
                     [ord(primera[2]), ord(segunda[2]), ord(tercera[2])]])

ctoi = lambda x: x - ord('A')
encriptado
```

```python
encriptado = encriptado.applyfunc(ctoi)
encriptado
```

```python
encriptado = encriptado.inv_mod(26)
encriptado
```

```python
posibles = []
permutaciones = permutations(freqTri, 3)
for permutacion in permutaciones:
    primera, segunda, tercera = permutacion
    original = Matrix([[ord(primera[0]), ord(segunda[0]), ord(tercera[0])],
       [ord(primera[1]), ord(segunda[1]), ord(tercera[1])],
       [ord(primera[2]), ord(segunda[2]), ord(tercera[2])]])
    original = original.applyfunc(ctoi)

    if original.det() != 0:
        resultado = (original * encriptado) % 26
        
        correcto = True
        for i in resultado:
            if type(i) == Rational or type(i) == sp.core.numbers.Half:
                correcto = False
                break
        if correcto:
            posibles.append(resultado)
```

```python
print(len(posibles))
print(posibles)
```

```python
trocito = texto[0:3*20]
print(len(trocito))
print(trocito)
```

```python
for matriz in posibles:
    linia = ""
    for i in range(3, len(trocito) + 3, 3):
        palabra = trocito[i-3:i]
        primera, segunda, tercera = palabra[0], palabra[1], palabra[2]
        leido = Matrix([ord(primera), ord(segunda), ord(tercera)])
        leido = leido.applyfunc(ctoi)
        
        resultado = matriz * leido
        resultado = resultado % 26
        
        for j in resultado:
            linia += chr(ord('A') + j)
    
    print(linia[0:10])
```

```python
matriz = posibles[0]
trocito = texto
linia = ""
for i in range(3, len(trocito) + 3, 3):
    palabra = trocito[i-3:i]
    primera, segunda, tercera = palabra[0], palabra[1], palabra[2]
    leido = Matrix([ord(primera), ord(segunda), ord(tercera)])
    leido = leido.applyfunc(ctoi)

    resultado = matriz * leido
    resultado = resultado % 26

    for j in resultado:
        linia += chr(ord('A') + j)

print(linia)
```
