La funcion enunciado le pasas una lista con los exponentes y te pone directamente el polinomio

```python
def enunciado(exponentes):
    final = ""
    for i in range(15, -1, -1):
        if i in exponentes:
            final += "1"
        else:
            final += "0"
    return final

def polinomio(n):
    binario = bin(n)[2:]
    binario = '0'*16 + binario
    binario = binario[-16:]
    return binario

def grado(n):
    for index, element in enumerate(n):
        if element == '1':
            return 15 - index
    return 0

def shift(n, a):
    n = int(n, 2)
    n = n << a
    return polinomio(n)

def xor(m, n):
    m = int(m, 2)
    n = int(n, 2)
    return polinomio(m ^ n)

def divisible(m, n):
    m = polinomio(m)
    n = polinomio(n)
    
    gradom = grado(m)
    gradon = grado(n)
    
    while gradom >= gradon:
        dif = gradom - gradon
        m = xor(shift(n, dif), m)
        gradom = grado(m)
        
    return int(m, 2) == 0

def irreductible_polynomials():
    irreductibles = []
    for i in range(2, 512):
        if len(irreductibles) == 0:
            irreductibles.append(i)
        else:
            primo = True
            for j in irreductibles:
                primo = not divisible(i, j)
                if not primo:
                    break
            if primo:
                irreductibles.append(i)
    return irreductibles

def grados(n):
    grados = []
    for index, element in enumerate(n):
        if element == "1":
            grados.append(15 - index)
    return grados

def division(m, n):
    gradom = grado(m)
    gradon = grado(n)
    
    while gradom >= gradon:
        dif = gradom - gradon
        m = xor(shift(n, dif), m)
        gradom = grado(m)
        
    return m

def GF_product_p(a, b):
    a = polinomio(a)
    b = polinomio(b)
    gradosl = grados(b)
    polinomiosl = []
    for grado in gradosl:
        if grado == 0:
            polinomiosl.append(a)
        else:
            polinomiosl.append(shift(a, grado))
    
    result = None
    for i in polinomiosl:
        if result is None:
            result = i
        else:
            result = xor(result, i)
    
    if result is None:
        return 0
    
    else:
        result = division(result, irreductible)
        return int(result, 2)
    
def GF_es_generador(a):
    tabla = set()
    tabla.add(1)
    anterior = 1
    for i in range(1, 255):
        anterior = GF_product_p(anterior, a)
        tabla.add(anterior)
    else:
        return len(tabla) == 255
    
def generador_minimo():
    generador = None
    for i in range(0, 255):
        if GF_es_generador(i):
            generador = i
            break
    return generador

def GF_tables():
    exponencial = [1] * 255
    logaritmica = [0] * 256
    for i in range(1, 255):
        result = GF_product_p(exponencial[i - 1], generador_minimo())
        logaritmica[result] = i
        exponencial[i] = result
    return exponencial, logaritmica

def potencia(a, e):
    result = 1
    for i in range(0, e):
        result = GF_product_p(result, a)
    return result

def GF_product_t(a, b):
    if a == 0 or b == 0:
        return 0
    exponente1 = logaritmica[a]
    exponente2 = logaritmica[b]
    exponenteT = (exponente1 + exponente2) % 255
    return exponencial[exponenteT]

def GF_invers(a):
    exponente = logaritmica[a]
    return exponencial[(255 - exponente) % 255]
```
