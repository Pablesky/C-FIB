Cambiar *rutaOriginal* por el texto deseado.

Interpreta que el carácter mas usado es la e, cambiar *mostFreq*.

```python
rutaOriginal = 'CesarExamen'
fichero = open(rutaOriginal)
texto = fichero.read()
fichero.close
print(texto[0:20])
```

```python
frecuencias = {}
for char in texto:
    if char in frecuencias.keys():
        frecuencias[char] += 1
    else:
        frecuencias[char] = 0
print(frecuencias)
```

```python
sortedD = {}
sortedK = sorted(frecuencias, key=frecuencias.get, reverse=True)  # [1, 3, 2]

for w in sortedK:
    sortedD[w] = frecuencias[w]

frecuencias = sortedD
print(frecuencias)
```

```python
llaves = list(frecuencias.keys())
if llaves[0] == ' ':
    frecuencias.pop(' ')
print(frecuencias)
```

```python
mostFreq = 'e'
offset = (ord(llaves[0]) - ord(mostFreq)) % 26
print(offset)
```

```python
unaLinia = ""
for i in texto:
    nuevo = chr(ord('a') + (ord(i) - offset) % 26)
    if not i in [' ', ',', '.', '\n']:
        unaLinia += nuevo
    else:
        unaLinia += i
print(unaLinia[:10])
```
