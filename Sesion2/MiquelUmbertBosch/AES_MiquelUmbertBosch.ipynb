# Desencripta b√°sico

```python
from Crypto.Cipher import AES
```

```python
key = open('AES_miquel.umbert.bosch_2022_09_29_11_10_42.key', 'rb').read()
print(key)
print(len(key))
```

```python
iv = open('AES_miquel.umbert.bosch_2022_09_29_11_10_42.iv', 'rb').read()
print(iv)
print(len(iv))
```

```python
texto = open('AES_miquel.umbert.bosch_2022_09_29_11_10_42.enc', 'rb').read()
print(len(texto))
```

```python
obj = AES.new(key, AES.MODE_CBC, iv)
algo = obj.decrypt(texto)

import filetype
tipo = filetype.guess(algo)
print(tipo)
```

```python
import sys
from PIL import Image
from io import BytesIO

stream = BytesIO(algo)

image = Image.open(stream).convert("RGBA")
stream.close()
image.show()
```

# Desencripta Avanzado

```python
import hashlib
m = hashlib.sha256()
print(b"hola")
m.update(b"hola")
m.digest()

encriptado = iv = open('AES_miquel.umbert.bosch_2022_09_20_16_59_35.puerta_trasera.enc', 'rb').read()
```

```python
import filetype

tipos = []
listak = []
listav = []

for primero in range(0, 2**7):
    for segundo in range(0, 2**7):
        print(primero, segundo)
        letras = chr(primero)*8 + chr(segundo)*8
        m = hashlib.sha256()
        m.update(letras.encode('UTF-8'))
        premasterkey = m.digest()
        key = premasterkey[0:16]
        iv = premasterkey[16:32]
        
        obj = AES.new(key, AES.MODE_CBC, iv)
        desencriptado = obj.decrypt(encriptado)
        tipo = filetype.guess(desencriptado)
        if not tipo is None:
            tipos.append(tipo)
            listak.append(key)
            listav.append(iv)
            print(str(tipo))
            
```

```python
print(tipos)
```

```python
print(tipos[0])
key = listak[0]
iv = listav[0]
obj = AES.new(key, AES.MODE_CBC, iv)
desencriptado = obj.decrypt(encriptado)
desencriptado
```

```python
with open("videoMiquelUmbertBosch.mp4", "wb") as binary_file:
   
    # Write bytes to file
    binary_file.write(desencriptado)
```
